## **Detecting Outliers**

### **Visualizing Continuous Variables with Boxplots**

```python
#Identify continuous variables (assuming all numerical columns are continuous)
continuous_variables = df.select_dtypes(include=['int', 'float']).columns

# Plot boxplots for each continuous variable
for variable in continuous_variables:
    sns.boxplot(x=df[variable])
    plt.title(f'Boxplot for {variable}')
    plt.show()
```

### **Handling Outliers for Continuous Variables**

```python
# Remove or clip outliers for each continuous variable
for variable in continuous_variables:
    lower_bound = df[variable].quantile(0.05)
    upper_bound = df[variable].quantile(0.95)
    df[variable] = np.clip(df[variable], lower_bound, upper_bound)

# Visualize the distribution of each continuous variable using boxplot
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('Boxplot of Continuous Variables after Clipping Outliers')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

### **Analysis of Marginal Probability: Customer Preferences for Product Purchases**

```python
# Calculate the marginal probability of each product
marginal_probability = df['Product'].value_counts(normalize=True) * 100

# Display the marginal probability
print("Marginal Probability of Each Product:")
print(marginal_probability)

# Plot the bar plot
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green
plt.figure(figsize=(8, 6))
marginal_probability.plot(kind='bar', color=colors)
plt.title('Marginal Probability of Each Product')
plt.xlabel('Product')
plt.ylabel('Probability (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
#Calculate the probability that a customer buys a product based on each column
probability_based_on_column = {}

# Iterate over each column in the crosstab
for column in marginal_probability.columns:
    # Calculate the probability based on the marginal probability of each product
    probability_based_on_column[column] = marginal_probability.loc[:, column] / marginal_probability.loc[:, column].sum()

# Display the probabilities
print("Probability that the customer buys a product based on each column:")
print(probability_based_on_column)


# Plot the probabilities
plt.figure(figsize=(10, 6))
for i, (column, probability) in enumerate(probability_based_on_column.items()):
    plt.bar(probability.index, probability.values, label=column, color=colors[i])

plt.xlabel('Product')
plt.ylabel('Probability (%)')
plt.title('Probability of Customer Buying a Product Based on Each Column')
plt.legend(title='Column')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### **üîç Insights**

The marginal probability represents the percentage of customers who have purchased each product. Here are the insights based on the marginal probability of each product:

1. **KP281**: Approximately 44.44% of customers have purchased KP281, making it the most popular product among the three.
  
2. **KP481**: Around 33.33% of customers have purchased KP481, indicating a significant portion of the customer base prefers this product.
  
3. **KP781**: About 22.22% of customers have purchased KP781, making it the least purchased product among the three.

These insights provide a clear understanding of the distribution of product purchases among customers, helping in strategic decision-making for product marketing and inventory management.

## **Analysis of Gender-Based Conditional Probability in Product Purchases**

```python
# Get unique product names
products = df['Product'].unique()

# Calculate the conditional probability for each product given each gender
conditional_probabilities = {}
for product in products:
    product_conditional_probabilities = {}
    for gender in df['Gender'].unique():
        event_A_and_B = len(df[(df['Gender'] == gender) & (df['Product'] == product)])
        event_B = len(df[df['Gender'] == gender])
        conditional_probability = event_A_and_B / event_B
        product_conditional_probabilities[gender] = conditional_probability
    conditional_probabilities[product] = product_conditional_probabilities

# Display the conditional probabilities
for product, probabilities in conditional_probabilities.items():
    print(f"Conditional Probability (Purchasing {product}):")
    for gender, probability in probabilities.items():
        print(f"{gender}: {probability}")
    print()

df_conditional = pd.DataFrame(conditional_probabilities)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_conditional, annot=True, cmap='Blues', fmt=".2f")  # Previous colormap 'Blues'

# Add labels and title
plt.xlabel('Product')
plt.ylabel('Gender')
plt.title('Conditional Probability of Purchasing Each Product by Gender')

# Show plot
plt.show()
```

#### **üîç Insights**
- **KP281**: Female customers exhibit a higher likelihood (52.63%) of purchasing KP281 compared to male customers (38.46%).
- **KP481**: Both male and female customers show a preference for KP481, with probabilities of 29.81% and 38.16%, respectively.
- **KP781**: Male customers demonstrate a slightly higher likelihood (31.73%) of purchasing KP781 compared to female customers (9.21%).

  This summary highlights the varying probabilities of product purchases between genders.
